```python
from collections import defaultdict

with open('../result_clf.json', 'r') as f:
    result = json.load(f)

d_bboxes = defaultdict(list)
d_scores = defaultdict(list)
d_categories = defaultdict(list)
d_clfs = defaultdict(list)
d_objs = defaultdict(list)

for anno in result:
    id_ = anno['image_id']
    d_bboxes[id_].append([anno['bbox'][0], anno['bbox'][1], anno['bbox'][0] + anno['bbox'][2], anno['bbox'][1] + anno['bbox'][3]])
    d_scores[id_].append(anno['score'])
    d_categories[id_].append(anno['category_id'])
    d_clfs[id_].append(anno['clf'])
    d_objs[id_].append(anno['obj'])
    
for x in d_bboxes:
    d_bboxes[x] = np.array(d_bboxes[x])
    d_scores[x] = np.array(d_scores[x])
    d_categories[x] = np.array(d_categories[x])
    d_clfs[x] = np.array(d_clf[x])
    d_objs[x] = np.array(d_objs[x])
    sort_idx = np.argsort(d_scores[x])[::-1]
    d_bboxes[x] = d_bboxes[x][sort_idx]
    d_scores[x] = d_scores[x][sort_idx]
    d_categories[x] = d_categories[x][sort_idx]
    d_clfs[x] = d_clfs[x][sort_idx]
    d_objs[x] = d_objs[x][sort_idx]
```

```python
from collections import defaultdict

eval_range = [0, 10000000000.0]

FP = defaultdict(list)
TP = defaultdict(list)

for img in cocoEval.evalImgs:

    if img is None or img['aRng'] != eval_range:
        continue

    label = img['category_id']

    for det_id, flag in zip(img['dtIds'], img['dtMatches'][0]):
        
        det_info = cocoDt.loadAnns(det_id)[0]
        det_score = det_info['score']
        det_area = det_info['area']
        
        # if det_score < score_threshold:
        #     continue
        
        if flag == 0:
            FP[label].append([det_score, det_area])
        else:
            TP[label].append([det_score, det_area])

assert sum(len(l) for l in FP.values()) + sum(len(l) for l in TP.values()) == len(cocoDt.anns)
```

```python
import json

with open('../dataset/train.json') as json_file:
    json_data = json.load(json_file)
```

```python
from sklearn.model_selection import train_test_split

train, val = train_test_split(json_data['images'], test_size=0.2, random_state=42)
```

```python
import copy

json_train = copy.deepcopy(json_data)
json_val = copy.deepcopy(json_data)

json_train['images'] = train
json_val['images'] = val
```

```python
imgs = set(d['id'] for d in json_train['images'])
json_train['annotations']= [d for d in json_train['annotations'] if d['image_id'] in imgs]

imgs = set(d['id'] for d in json_val['images'])
json_val['annotations']= [d for d in json_val['annotations'] if d['image_id'] in imgs]
```

```python
with open("train_split.json", "w") as json_file:
    json.dump(json_train, json_file)

with open("val_split.json", "w") as json_file:
    json.dump(json_val, json_file)
```
